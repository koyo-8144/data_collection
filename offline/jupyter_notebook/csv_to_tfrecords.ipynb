## Filter out joint states to align with the number of end-effector states

## Sanity check: Calculate mean time discrepancy between joint and EEF states

```python

def to_unix_timestamp(datetime_string):
    datetime_string = datetime_string[:-3]  # remove last 2 microseconds, datetime supports only 6
    return datetime.datetime.strptime(datetime_string, "%Y/%m/%d %H:%M:%S.%f").timestamp()

ee_times = ee_states_aligned_df['time'].apply(to_unix_timestamp).to_list()
ee_times = np.asarray(ee_times)

joint_times = joint_states_aligned_df['time'].apply(to_unix_timestamp).to_list()
joint_times = np.asarray(joint_times)

mean_time_s = abs(ee_times - joint_times).mean()
print(f'Mean time misalignment between joints and EEF states is: {mean_time_s:.3f} seconds')

# make sure the time difference is not too big...
assert mean_time_s < 1  
```

```python
action_df = combined_df[['position_x', 'position_y', 'position_z', 'orientation_x', 'orientation_y', 'orientation_z']]
action_df = action_df.rename(columns={'position_x': 'delta_position_x', 
                          'position_y': 'delta_position_y', 
                          'position_z': 'delta_position_z', 
                          'orientation_x': 'delta_orientation_x', 
                          'orientation_y': 'delta_orientation_y', 
                          'orientation_z': 'delta_orientation_z'})

action_df = action_df[1:].reset_index(drop=True).copy() - action_df[:-1].reset_index(drop=True).copy()

assert action_df.shape[0] == combined_df.shape[0] - 1

print('Actions shape:', action_df.shape)

action_df.head()  # orientation_z is almost 0... 
```

```python
import tensorflow as tf
import csv
import os

print(tf.__version__)
```

```python
# Define header and feature details
HEADER = ['time', 'ee_state', 'joint_state']
HEADER_DEFAULTS = [['NA'], ['NA'], ['NA']]

NUMERIC_FEATURE_NAMES = ['ee_state', 'joint_state']
CATEGORICAL_FEATURE_NAMES = []
FEATURE_NAMES = NUMERIC_FEATURE_NAMES + CATEGORICAL_FEATURE_NAMES
TARGET_NAME = None  # If there's a target column, specify its name here
UNUSED_FEATURE_NAMES = list(set(HEADER) - set(FEATURE_NAMES) - {TARGET_NAME})

print("Header: {}".format(HEADER))
print("Numeric Features: {}".format(NUMERIC_FEATURE_NAMES))
print("Categorical Features: {}".format(CATEGORICAL_FEATURE_NAMES))
print("Unused Features: {}".format(UNUSED_FEATURE_NAMES))
```

```python
# Function to iterate through CSV rows
def create_csv_iterator(csv_file_path, skip_header):
    with open(csv_file_path, 'r') as csv_file:
        reader = csv.reader(csv_file)
        if skip_header:  # Skip the header
            next(reader)
        for row in reader:
            yield row
```

```python
# Function to create a tf.train.Example
def create_example(row):
    example = tf.train.Example()

    for i, feature_name in enumerate(HEADER):
        feature_value = row[i]

        if feature_name in UNUSED_FEATURE_NAMES:
            continue

        if feature_name in NUMERIC_FEATURE_NAMES:
            example.features.feature[feature_name].float_list.value.append(float(feature_value))

        elif feature_name in CATEGORICAL_FEATURE_NAMES:
            example.features.feature[feature_name].bytes_list.value.append(feature_value.encode('utf-8'))

    return example
```

```python
# Function to create TFRecords file
def create_tfrecords_file(input_csv_file):
    output_tfrecord_file = input_csv_file.replace(".csv", ".tfrecords")
    print("Creating TFRecords file at", output_tfrecord_file, "...")

    with tf.io.TFRecordWriter(output_tfrecord_file) as writer:
        for i, row in enumerate(create_csv_iterator(input_csv_file, skip_header=True)):
            if len(row) == 0:
                continue

            example = create_example(row)
            writer.write(example.SerializeToString())

    print("Finished writing", output_tfrecord_file)
```

```python
# Specify your CSV file
csv_file_path = "/home/koyo/data_collection/offline/bag_files/rosbag2_2024_11_25-19_13_17/rosbag2_2024_11_25-19_13_17_0.db/combined_df.csv"
data_files = [csv_file_path]

# Convert the CSV file
for input_csv_file in data_files:
    create_tfrecords_file(input_csv_file)
```
